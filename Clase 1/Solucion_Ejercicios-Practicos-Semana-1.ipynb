# PARTE 1: Consultas Básicas de Exploración

## Ejercicio 1.1: Tu primera query

**Objetivo:** Familiarizarte con Databricks y ejecutar tu primera query.  
**Pregunta:** ¿Cuántos viajes hay en total en el dataset de NYC Taxi Trips?  
**Pista:** Usa la función `COUNT(*)` sobre la tabla `samples.nyctaxi.trips`

```python
%sql
SELECT COUNT(*) as total_viajes
FROM samples.nyctaxi.trips;
```

## Ejercicio 1.2: Explorar estructura

**Objetivo:** Entender qué columnas tiene el dataset.  
**Pregunta:** ¿Qué columnas tiene la tabla `samples.nyctaxi.trips`? ¿Qué tipos de datos tienen?

```python
%sql
DESCRIBE samples.nyctaxi.trips;
```

## Ejercicio 1.3: Viajes más largos

**Objetivo:** Encontrar valores extremos en los datos.  
**Pregunta:** ¿Cuál es el viaje más largo (mayor distancia) en el dataset? ¿Y el más corto (excluyendo ceros)?  
**Pista:** Usa `MAX()` y `MIN()` con `WHERE` para filtrar valores inválidos

```python
%sql
SELECT
  MAX(trip_distance) as distancia_maxima,
  MIN(trip_distance) as distancia_minima
FROM samples.nyctaxi.trips
WHERE trip_distance > 0;
```

## Ejercicio 1.4: Estadísticas básicas

**Objetivo:** Calcular promedios y estadísticas descriptivas.  
**Pregunta:** ¿Cuál es el promedio de tarifa (`fare_amount`) y distancia (`trip_distance`) de los viajes?  
**Pista:** Usa `AVG()` y `ROUND()` para formatear decimales

```python
%sql
SELECT
  ROUND(AVG(fare_amount), 2) as tarifa_promedio,
  ROUND(AVG(trip_distance), 2) as distancia_promedio
FROM samples.nyctaxi.trips;
```

## Ejercicio 1.5: Viajes recientes

**Objetivo:** Ordenar y limitar resultados.  
**Pregunta:** Muestra los 20 viajes más recientes con su fecha, distancia y tarifa.  
**Pista:** Usa `ORDER BY` con `DESC` y `LIMIT`

```python
%sql
SELECT
  tpep_pickup_datetime,
  trip_distance,
  fare_amount
FROM samples.nyctaxi.trips
ORDER BY tpep_pickup_datetime DESC
LIMIT 20;
```

# PARTE 4: GROUP BY

## Ejercicio 4.1: Agrupar por zona

**Objetivo:** Agrupar datos y calcular métricas por grupo.  
**Pregunta:** ¿Cuántos viajes hay por zona de inicio (`pickup_zip`)? ¿Cuál es el ingreso total y promedio por zona? Ordena por ingreso total descendente.  
**Pista:** Usa `GROUP BY pickup_zip`, `COUNT(*)`, `SUM(fare_amount)`, `AVG(fare_amount)`, `ORDER BY`

```python
%sql
SELECT
  pickup_zip,
  COUNT(*) as total_viajes,
  ROUND(SUM(fare_amount), 2) as ingresos_totales,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY ingresos_totales DESC;
```

## Ejercicio 4.2: Viajes por día de la semana

**Objetivo:** Agrupar por fecha y calcular estadísticas.  
**Pregunta:** ¿Cuántos viajes hay por día de la semana? ¿Cuál es la distancia promedio por día?  
**Pista:** Usa `DAYOFWEEK()` y `GROUP BY`

```python
%sql
SELECT
  DAYOFWEEK(tpep_pickup_datetime) as dia_semana,
  COUNT(*) as total_viajes,
  ROUND(AVG(trip_distance), 2) as distancia_promedio
FROM samples.nyctaxi.trips
WHERE trip_distance > 0
GROUP BY DAYOFWEEK(tpep_pickup_datetime)
ORDER BY dia_semana;
```

## Ejercicio 4.3: Viajes por hora

**Objetivo:** Analizar patrones horarios.  
**Pregunta:** ¿Cuántos viajes hay por hora del día? ¿Cuál es la tarifa promedio por hora? Ordena por hora.  
**Pista:** Usa `HOUR()` y `GROUP BY`

```python
%sql
SELECT
  HOUR(tpep_pickup_datetime) as hora,
  COUNT(*) as total_viajes,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio
FROM samples.nyctaxi.trips
GROUP BY HOUR(tpep_pickup_datetime)
ORDER BY hora;
```

# PARTE 5: HAVING

## Ejercicio 5.1: Filtrar grupos grandes

**Objetivo:** Filtrar grupos después de agrupar.  
**Pregunta:** Muestra solo las zonas que tienen más de 1,000 viajes.  
**Pista:** Usa `GROUP BY` con `HAVING COUNT(*) > 1000`

```python
%sql
SELECT
  pickup_zip as zona_inicio,
  COUNT(*) as total_viajes
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
HAVING COUNT(*) > 1000
ORDER BY total_viajes DESC;
```

## Ejercicio 5.2: Días con muchos viajes

**Objetivo:** Identificar días con alta actividad.  
**Pregunta:** ¿Qué días tuvieron más de 200 viajes? Muestra la fecha, cantidad de viajes y tarifa promedio.  
**Pista:** `GROUP BY`, `HAVING`, `ORDER BY` y `LIMIT`

```python
%sql
SELECT
  DATE(tpep_pickup_datetime) as fecha,
  COUNT(*) as total_viajes,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio
FROM samples.nyctaxi.trips
GROUP BY DATE(tpep_pickup_datetime)
HAVING COUNT(*) > 200
ORDER BY fecha DESC;
```

## Ejercicio 5.3: Horas pico

**Objetivo:** Identificar horas con mayor actividad.  
**Pregunta:** ¿Qué horas tienen más de 1,000 viajes? Ordena por cantidad de viajes descendente.  
**Pista:** `GROUP BY`, `HAVING COUNT(*) > 1000`

```python
%sql
SELECT
  HOUR(tpep_pickup_datetime) as hora,
  COUNT(*) as total_viajes,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio
FROM samples.nyctaxi.trips
GROUP BY HOUR(tpep_pickup_datetime)
HAVING COUNT(*) > 1000
ORDER BY total_viajes DESC;
```

# PARTE 6: Análisis Combinado

## Ejercicio 6.1: Análisis completo por zona

**Objetivo:** Combinar `WHERE`, `GROUP BY` y `HAVING`.  
**Pregunta:** Para zonas con más de 1,000 viajes, muestra: zona, cantidad de viajes, tarifa máxima, promedio y mínima. Solo incluye viajes con tarifa mayor a $5 y distancia mayor a 0.  
**Pista:** Usa `WHERE` para filtrar registros, `GROUP BY` para agrupar, `HAVING` para filtrar grupos

```python
%sql
SELECT
  pickup_zip,
  COUNT(*) as total_viajes,
  ROUND(MAX(fare_amount), 2) as tarifa_maxima,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio,
  ROUND(MIN(fare_amount), 2) as tarifa_minima
FROM samples.nyctaxi.trips
WHERE fare_amount > 5
  AND trip_distance > 0
GROUP BY pickup_zip
HAVING COUNT(*) > 1000
ORDER BY tarifa_promedio DESC;
```

## Ejercicio 6.2: Categorizar viajes por distancia

**Objetivo:** Usar `CASE WHEN` para crear categorías.  
**Pregunta:** Categoriza los viajes por distancia:  
- Corto: < 2 millas  
- Medio: 2-5 millas  
- Largo: 5-10 millas  
- Muy largo: > 10 millas  

Luego agrupa por categoría y muestra: categoría, cantidad de viajes, tarifa promedio, distancia promedio. Solo categorías con más de 100 viajes.  
**Pista:** Usa `CASE WHEN` en el `SELECT` y también en el `GROUP BY`

```python
%sql
SELECT
  CASE
    WHEN trip_distance < 2 THEN 'Corto (< 2 mi)'
    WHEN trip_distance BETWEEN 2 AND 5 THEN 'Medio (2-5 mi)'
    WHEN trip_distance BETWEEN 5 AND 10 THEN 'Largo (5-10 mi)'
    ELSE 'Muy largo (> 10 mi)'
  END as categoria,
  COUNT(*) as total_viajes,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio,
  ROUND(AVG(trip_distance), 2) as distancia_promedio
FROM samples.nyctaxi.trips
WHERE trip_distance > 0
  AND fare_amount > 0
GROUP BY
  CASE
    WHEN trip_distance < 2 THEN 'Corto (< 2 mi)'
    WHEN trip_distance BETWEEN 2 AND 5 THEN 'Medio (2-5 mi)'
    WHEN trip_distance BETWEEN 5 AND 10 THEN 'Largo (5-10 mi)'
    ELSE 'Muy largo (> 10 mi)'
  END
HAVING COUNT(*) > 100
ORDER BY tarifa_promedio DESC;
```
