```python
%sql

select pickup_zip, count(*) as total_trips,sum(fare_amount) as total_fare , avg(fare_amount) as avg_fare from trips
group by pickup_zip
order by sum(fare_amount) desc
```

```python
%sql

select dayofweek(tpep_pickup_datetime) as `Day of the week`,count(*) as `Total Trips`,avg(trip_distance) asavg_distance from trips
group by dayofweek(tpep_pickup_datetime)
```

```python
%sql

select hour (tpep_pickup_datetime) as hour,count(*) as `Total Trips`, avg(fare_amount) as  `avg fare` from trips
group by hour (tpep_pickup_datetime)
order by hour (tpep_pickup_datetime)
```

```python
%sql

use samples.nyctaxi;
select date(tpep_pickup_datetime), count(*) as `Total Trips`, avg(fare_amount) as `avg fare` from trips
group by date(tpep_pickup_datetime)
having count(*) > 200
order by date(tpep_pickup_datetime) desc
```

```python
%sql
use samples.nyctaxi;
select hour(tpep_pickup_datetime), count(*) as `Total Trips` from trips
group by hour(tpep_pickup_datetime)
having count(*) > 1000
order by count(*) desc
```

```python
%sql

select pickup_zip,count (*) as `Total trip`, max(fare_amount),avg(fare_amount),min(fare_amount) from trips
where fare_amount>5 and trip_distance >0
group by pickup_zip
having count(*) > 1000
order by count(*) desc
    
```

```python
%sql

select 
case when trip_distance < 2 then 'Corto'
when trip_distance between 2 and 5 then 'Medio'
when trip_distance between 5.01 and 10 then 'Largo'
else 'Muy largo' end as `Categoria`,
count (*) as `Total trips`,
avg(fare_amount) as `Avg fare`,
avg(trip_distance) as `Avg trip distance`
from trips
group by 
case when trip_distance < 2 then 'Corto'
when trip_distance between 2 and 5 then 'Medio'
when trip_distance between 5.01 and 10 then 'Largo'
else 'Muy largo' end
having count(*) >100
```
