### Ejercicio 1.1: Variables, tipos y operaciones — SETUP

```python
# Variables de una propiedad inmobiliaria
zona = "Palermo"
precio = 185000
moneda = "USD"
metros = 92.5
ambientes = 3
tiene_cochera = True

# type() — verificar tipo de cada variable
print(f"zona:          {type(zona)}")
print(f"precio:        {type(precio)}")
print(f"metros:        {type(metros)}")
print(f"tiene_cochera: {type(tiene_cochera)}")

# Operaciones aritméticas
precio_por_m2 = round(precio / metros, 2)
print(f"\nPrecio por m2: USD {precio_por_m2}")
print(f"Precio en miles: {precio // 1000}K")
print(f"Resto div 1000: {precio % 1000}")

# String operations
print(f"\nZona mayúscula: {zona.upper()}")
print(f"Zona minúscula: {zona.lower()}")
print(f"Largo: {len(zona)} caracteres")
direccion = "Av. Santa Fe 3200"
print(f"Primeros 2: {direccion[:2]}")
print(f"Últimos 4: {direccion[-4:]}")
print(f"Reemplazo: {direccion.replace('Av.', 'Avenida')}")

# Bool — comparaciones
print(f"\n¿Precio > 100K? {precio > 100000}")
print(f"¿3+ ambientes y cochera? {ambientes >= 3 and tiene_cochera}")
print(f"¿Barato o nuevo? {precio < 50000 or False}")
print(f"¿No tiene cochera? {not tiene_cochera}")

# Conversiones de tipo
precio_texto = "250000"
print(f"\nstr→int: {int(precio_texto) + 10000}")
print(f"float→int: {int(metros)}")
print(f"int→str: {'El precio es ' + str(precio)}")
```

### Ejercicio 1.2: f-strings para reporte — SETUP

```python
# f-string: ficha individual
p = {"zona": "Palermo", "precio": 185000, "moneda": "USD", "metros": 92.5}
print(f"Propiedad en {p['zona']}: {p['moneda']} {p['precio']:,} — {p['metros']}m2")
print(f"Precio por m2: {p['moneda']} {p['precio'] / p['metros']:,.2f}")

# f-string: tabla alineada
propiedades = [
    {"zona": "Palermo", "precio": 185000, "metros": 92.5},
    {"zona": "Belgrano", "precio": 320000, "metros": 145.0},
    {"zona": "Caballito", "precio": 78000, "metros": 55.0},
    {"zona": "Tigre", "precio": 250000, "metros": 200.0},
]

print(f"\n{'Zona':<15} {'Precio':>12} {'Metros':>8} {'USD/m2':>10}")
print("-" * 47)
for prop in propiedades:
    pm2 = round(prop["precio"] / prop["metros"], 2)
    print(f"{prop['zona']:<15} {prop['precio']:>12,} {prop['metros']:>8.1f} {pm2:>10,.2f}")
```

### Ejercicio 1.3: list + set + dict — GUIDED


```python
# LIST — colección ordenada
precios = [185000, 320000, 78000, 250000, 145000, 320000]

print("=== LIST ===")
print(f"Cantidad: {len(precios)}")
print(f"Suma: {sum(precios):,}")
print(f"Promedio: {sum(precios) / len(precios):,.2f}")
print(f"Máximo: {max(precios):,}")
print(f"Mínimo: {min(precios):,}")
print(f"Primero: {precios[0]:,} | Último: {precios[-1]:,}")

precios.append(410000)
print(f"\nDespués de append: {len(precios)} elementos")
print(f"Ordenado asc: {sorted(precios)}")
print(f"Ordenado desc: {sorted(precios, reverse=True)}")
```

```python
# SET — sin duplicados (equivalente a SELECT DISTINCT)
zonas_todas = ["Palermo", "Belgrano", "Palermo", "Caballito", "Belgrano", "Tigre", "Palermo"]
zonas_unicas = set(zonas_todas)

print("=== SET ===")
print(f"Lista original ({len(zonas_todas)}): {zonas_todas}")
print(f"Set único ({len(zonas_unicas)}): {zonas_unicas}")

zonas_caba = {"Palermo", "Belgrano", "Caballito"}
zonas_gba = {"Tigre", "San Isidro", "Vicente López"}
print(f"\nUnión:        {zonas_caba | zonas_gba}")
print(f"Intersección: {zonas_caba & zonas_gba}")
print(f"Solo CABA:    {zonas_caba - zonas_gba}")
```

```python
# DICT — clave:valor (una fila de tabla)
propiedad = {
    "zona": "Palermo",
    "precio": 185000,
    "moneda": "USD",
    "metros": 92.5,
    "ambientes": 3
}

print("=== DICT ===")
print(f"Zona: {propiedad['zona']}")
print(f"Keys: {list(propiedad.keys())}")
print(f"Values: {list(propiedad.values())}")

# Agregar clave calculada
propiedad["precio_por_m2"] = round(propiedad["precio"] / propiedad["metros"], 2)
print(f"\nCon precio_por_m2: {propiedad['precio_por_m2']} USD/m2")

# Iterar con .items()
print("\nContenido completo:")
for clave, valor in propiedad.items():
    print(f"  {clave}: {valor}")
```

### Ejercicio 1.4: for loop + función — GUIDED


```python
# Función con edge case
def calcular_precio_m2(precio, metros):
    if metros == 0:
        return None
    return round(precio / metros, 2)

# Lista de propiedades como dicts
propiedades = [
    {"zona": "Palermo", "precio": 185000, "metros": 92.5},
    {"zona": "Belgrano", "precio": 320000, "metros": 145.0},
    {"zona": "Caballito", "precio": 78000, "metros": 55.0},
    {"zona": "Terreno vacío", "precio": 45000, "metros": 0},
]

print(f"{'#':<4} {'Zona':<15} {'Precio':>10} {'Metros':>8} {'USD/m2':>10}")
print("-" * 49)
for i, prop in enumerate(propiedades, start=1):
    pm2 = calcular_precio_m2(prop["precio"], prop["metros"])
    pm2_str = f"{pm2:,.2f}" if pm2 is not None else "N/A"
    print(f"{i:<4} {prop['zona']:<15} {prop['precio']:>10,} {prop['metros']:>8.1f} {pm2_str:>10}")

# range() para repetir N veces
print("\nConteo regresivo:")
for i in range(3, 0, -1):
    print(f"  {i}...")
```

### Ejercicio 1.7: Decision Framework — INDEPENDENT


```python
from pyspark.sql.functions import col, avg, count

# Forma 1: DataFrame API
resultado_df = (
    spark.table("bootcamp.silver.propiedades")
    .filter(col("tipo_operacion") == "Venta")
    .groupBy("partido")
    .agg(
        count("*").alias("cantidad"),
        avg("precio").alias("precio_promedio")
    )
)
resultado_df.show(5)
```

```python
# Forma 2: spark.sql()
resultado_sql = spark.sql("""
    SELECT partido, COUNT(*) as cantidad, AVG(precio) as precio_promedio
    FROM bootcamp.silver.propiedades
    WHERE tipo_operacion = 'Venta'
    GROUP BY partido
""")
resultado_sql.show(5)
```

```python
%sql
-- Forma 3: SQL puro
SELECT partido, COUNT(*) as cantidad, AVG(precio) as precio_promedio
FROM bootcamp.silver.propiedades
WHERE tipo_operacion = 'Venta'
GROUP BY partido
LIMIT 5;
```
